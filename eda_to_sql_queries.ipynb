```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
# Load SQL extension for using magic commands
%load_ext sql
```

```python
# Connect to the database
db_path = r'TMDB_movie_dataset_v11.db'

conn = sqlite3.connect(db_path)

#Query the database to get all data from table 'kaggle_movie'
query = "SELECT * FROM movie"

movie_df = pd.read_sql(query, conn)

conn.close()
```

```python
conn = sqlite3.connect(db_path)

#Query the database to get all data from table 'kaggle_movie'
query = "SELECT * FROM movie Where vote_count >= 25000"

movie_count = pd.read_sql(query, conn)

conn.close()
```

```python
conn = sqlite3.connect(db_path)

#Query the database to get all data from table 'kaggle_movie'
query = "SELECT DISTINCT title FROM movie"

unique_movie = pd.read_sql(query, conn)

conn.close()
```
